```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rf_speeds = [25.42,33.40,42.00]
xgboost_speeds = [84.00,120.00,142.8]
lightgbm_speeds = [3.42,3.64,3.72]

df = pd.DataFrame()
df['Random Forest'] = rf_speeds
df['XGBoost'] = xgboost_speeds
df['lightGBM'] = lightgbm_speeds
df = df.T
df.columns = ['Top20','Top25','Top30']
df
```

```python
plt.figure(figsize=(10,5),dpi=300)
sns.lineplot(data=df.T,marker='o',linewidth=2.5,markersize=10,palette='Set2')
plt.title('Speeds of Different Models')
plt.xlabel('Models')
plt.ylabel('Speeds (s)')
plt.savefig('speeds.png',dpi=300)
plt.show()
```
