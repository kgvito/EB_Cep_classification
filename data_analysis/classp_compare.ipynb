```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lightgbm_classp = [94.72,94.94,94.34,94.18,93.71,93.79]
rf_classp = [94.50,94.87,94.15,94.19,93.67,93.94]
xgboost_classp = [94.63,94.52,94.40,94.33,94.07,94.04]

df = pd.DataFrame()
df['Random Forest'] = rf_classp
df['XGBoost'] = xgboost_classp
df['lightGBM'] = lightgbm_classp
df = df.T
df.columns = ['>95','>90','>80','>60','>40','>20']
df
```

```python
plt.figure(figsize=(10,5),dpi=300)
sns.lineplot(data=df.T,marker='o',linewidth=2.5,markersize=10,palette='Set2')
plt.title('Speeds of Different Models')
plt.xlabel('Sample label confidence (%)')
plt.ylabel('Classification accuracy (weighted average) (%)')
plt.savefig('classp.png',dpi=300)
plt.show()
```
